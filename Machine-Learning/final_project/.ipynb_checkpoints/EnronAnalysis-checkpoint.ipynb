```python
# I am immediately going to add engineered variables
# I assume poi's have relatively greater email exchange with other pois
# Create percentage of poi-related email to total emails for each individual
pdata['poi_mail'] = ( pdata['from_poi_to_this_person'] + pdata['from_this_person_to_poi'] + pdata['shared_receipt_with_poi'] ) / ( pdata['to_messages'] + pdata['from_messages'] )

# I assume that people receiving relatively greater incentive pay are receiving it for a reason
# Create percentage of incentive pay to salary
pdata['s_ratio'] = ( pdata['bonus'] ) / pdata['salary']
# pdata['total_stock_value']
```

```python
# Extract features and labels from dataset for local testing
my_dataset = pdict
data = featureFormat(my_dataset, features_list, sort_keys = True)
labels, features = targetFeatureSplit(data)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
features = scaler.fit_transform(features)
```

```python
# Split data into training and test set
from sklearn.cross_validation import train_test_split
features_train, features_test, labels_train, labels_test = \
    train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
# Start with Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.grid_search import GridSearchCV
rfc = RandomForestClassifier(n_estimators=100)
param_grid = {"max_depth": [3, None],
              "max_features": [1, 3, 6],
              "min_samples_split": [1, 3, 10],
              "min_samples_leaf": [1, 3, 10],
              "bootstrap": [True, False],
              "criterion": ["gini", "entropy"]}
rfc_clf = GridSearchCV(rfc, param_grid=param_grid)
rfc_clf.fit(features_train, labels_train)
print rfc_clf.best_estimator_ 
```

```python
# rfc = RandomForestClassifier(bootstrap=False, min_samples_leaf=3, 
#                              min_samples_split=1, criterion='entropy', 
#                              max_features=1, max_depth=None)
from sklearn import cross_validation
kfold = cross_validation.KFold(len(features), n_folds=10)
xv_list = [rfc.fit(features_train, labels_train).score(features_test, labels_test) for train, test in kfold]
sum(xv_list) / len(xv_list)
# rfc = rfc.fit(features_train, labels_train)
# rfc
```

```python
# Look at relative feature importances
rfc.feature_importances_
```

```python
# Make predictions of poi on test set with random forest classifier
rfc_preds = rfc.predict(features_test)
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(labels_test, rfc_preds)
print(accuracy)
```

```python
# 12.5% of people in data are POIs. So predicting everyone is NOT POI would yield
# an accuracy score of 87.5%. Accuracy of 93.0% is marginal improvement.
```

```python
# But accuracy is only half the story. What about precision and recall...
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
for i in range(10):
    rfc = RandomForestClassifier(bootstrap=False, min_samples_leaf=3, min_samples_split=1, criterion='entropy', max_features=1, max_depth=None)
    rfc.fit(features_train, labels_train)
    rfc_preds = rfc.predict(features_test)
    cm = confusion_matrix(labels_test, rfc_preds)
    ps = precision_score(labels_test, rfc_preds)
    rs = recall_score(labels_test, rfc_preds)
    print(cm)
    print('ps: ', ps, ' rs: ', rs)
```
